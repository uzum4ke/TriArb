```python
import ccxt # for accessing crypto exchanges
import json # for saving data
from tqdm import tqdm # progress bar
```

STEP 0

```python
# get all the exchanges
def get_coin_pairs(exchanges):
    for exchange in exchanges:
        if exchange == "poloniex":
            exchange = ccxt.poloniex()


        pairs = exchange.load_markets()
        tradable_coins = []

        for pair in pairs:
            if pairs[pair]['active'] == True:
                tradable_coins.append(pair)
                
        print(f'{exchange.id} tradable currency pairs: {len(tradable_coins)}')

        return tradable_coins
```

STEP 1

```python
# get all the pairs that are triangular arbitrageable
def structure_triangular_pairs(coin_list):
    
    triangular_pairs = []
    remove_duplicates = []
    pair_list = []
    
    # Get pair A
    for pair_a in tqdm(coin_list):
        split_pair_a = pair_a.split("/")
        a_base = split_pair_a[0]
        a_quote = split_pair_a[1]
        a_pair_box = [a_base, a_quote]

        # Get pair B 
        for pair_b in coin_list:
            split_pair_b = pair_b.split("/")
            b_base = split_pair_b[0]
            b_quote = split_pair_b[1]
            b_pair_box = [b_base, b_quote]
            
            # Check if pair B is equal to pair A
            if pair_b != pair_a:
                if b_base in a_pair_box or b_quote in a_pair_box:

                    # Get pair C
                    for pair_c in coin_list:
                        split_pair_c = pair_c.split("/")
                        c_base = split_pair_c[0]
                        c_quote = split_pair_c[1]

                        # Check if pair C is equal to pair A or pair B
                        if pair_c != pair_a and pair_c != pair_b:
                            combine_all = [pair_a, pair_b, pair_c]
                            pair_box = [a_base, a_quote, b_base, b_quote, c_base, c_quote]
                            
                            # Remove duplicates
                            if ((c_base == b_base or c_base == b_quote) and (c_quote == a_base or c_quote == a_quote)) \
                                or ((c_quote == b_base or c_quote == b_quote) and (c_base == a_base or c_base == a_quote)):
                                
                                if sorted(combine_all) not in remove_duplicates:
                                    remove_duplicates.append(sorted(combine_all))
                                    pair_list.append([pair_a, pair_b, pair_c])


    return pair_list
```

```python
exchanges = ['poloniex']

tradable_coins = get_coin_pairs(exchanges)

structured_list = structure_triangular_pairs(tradable_coins)
print(len(structured_list))
```
